```python
import sqlite3
import requests
import datetime
# 有两个参数一个是限制更新多少行， 一个是上次更新多少天以内的就跳过，如果写0就是全更新
# API endpoint and headers
url = "https://seeking-alpha.p.rapidapi.com/symbols/get-estimates"
headers = {
    "x-rapidapi-key": "2716f4c647msh2f6be98768e11d3p1bc7a8jsnee05d6c307eb",
    "x-rapidapi-host": "seeking-alpha.p.rapidapi.com"
}

# Database path
db_path = "plan.db"

# Limit the number of updates
update_limit = 100

# Days limit for skipping updates 三十天 以内更新的跳过
days_limit = 25  # Can be modified as needed

# Fetch data from API and update database
def update_database(symbol):
    querystring = {
        "symbol": symbol,
        "data_type": "eps",
        "period_type": "annual"
    }

    response = requests.get(url, headers=headers, params=querystring)

    if response.status_code != 200:
        print(f"API Request Failed for {symbol}. Status Code: {response.status_code}")
        return False

    data = response.json().get('data', [])

    if not data:
        print(f"No data found for {symbol}.")
        return False

    eps_25, eps_26, period_25, period_26 = None, None, None, None

    for item in data:
        attributes = item.get('attributes', {})
        year = attributes.get('year')
        consensus_eps = attributes.get('consensus')
        period_end_date = attributes.get('period_end_date')

        if year == 2025:
            eps_25 = consensus_eps
            period_25 = period_end_date
        elif year == 2026:
            eps_26 = consensus_eps
            period_26 = period_end_date

    if eps_25 is None or eps_26 is None:
        print(f"Incomplete data for {symbol}. Skipping update.")
        return False

    try:
        with sqlite3.connect(db_path) as conn:
            cursor = conn.cursor()
            yoy_growth = 100*(eps_26 - eps_25) / eps_25 if eps_25 else None
            current_date = datetime.date.today().strftime('%Y-%m-%d')

            cursor.execute(
                """
                UPDATE CValue
                SET EPSY = ?,
                    FEPS = ?,
                    FFEPS = ?,
                    FEPSupdate = ?,
                    YOYGrowth = ?
                WHERE Symbol = ?;
                """,
                (period_25, eps_25, eps_26, current_date, yoy_growth, symbol)
            )
            conn.commit()
            print(f"Successfully updated {symbol}.")
            return True
    except Exception as e:
        print(f"Database update failed for {symbol}: {e}")
        return False

# Main program
if __name__ == "__main__":
    try:
        with sqlite3.connect(db_path) as conn:
            cursor = conn.cursor()
            cursor.execute("SELECT Symbol, FEPSupdate,ER_Date FROM CValue;")
            rows = cursor.fetchall()

            updates_count = 0
            today = datetime.date.today()

            for row in rows:
                if updates_count >= update_limit:
                    print("Reached update limit. Stopping.")
                    break

                symbol, feps_update,er_date = row
                if feps_update:
                    try:
                        # 这个条件是30天内没更新了，同时恰巧是在财报7-14天内就更新
                        feps_update_date = datetime.datetime.strptime(feps_update, "%Y-%m-%d").date()
                        er_date= datetime.datetime.strptime(er_date, "%m/%d/%Y").date()
                        if (today - feps_update_date).days <= days_limit:
                            print(f"Skipping {symbol}: Updated within {days_limit} days.")
                            continue
                        elif (today - er_date).days < 7 or (today - er_date).days > 14:
                            # 跳过：ER 日期距离今天在 7 天以内或 14 天以外
                            print("Skipping due to er_date being out of the 7-14 days range.")
                            #continue
                        else:    
                            print("Condition met, proceeding with the operation.")
                           
                    except ValueError:
                        print(f"Invalid date format for {symbol}: {feps_update}. Proceeding with update.")

                if update_database(symbol):
                    updates_count += 1

            print(f"Total updates performed: {updates_count}")
    except Exception as e:
        print(f"Error during database operation: {e}")
```
